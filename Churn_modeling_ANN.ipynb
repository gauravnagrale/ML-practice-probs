```python
x=pd.get_dummies(df.drop('Exited',axis=1),drop_first=True)
y=df['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from keras.models import Sequential
ann=Sequential()
```

```python
ann=tf.keras.models.Sequential()
from keras.layers import Dense,Dropout
```

# Adding the input layer and also the hidden layer

```python
ann.add(Dense(input_dim=11,units=6,activation='relu',kernel_initializer='uniform'))
# input_dim is the input that we have
# units is the no of the hidden layer 
# actiation activation function that we want to use
# Dense:-are the lines that we connect from the input layers to the hidden layer
ann.add(Dropout(rate=0.1))
# because we have overfitting problems we are dropping some Neurons randomly 
```

# Adding the second layer

```python
ann.add(Dense(units=6,activation='relu',kernel_initializer='uniform'))
ann.add(Dropout(rate=0.1))
```

# adding the OUTpUT layer

```python
ann.add(Dense(units=1,activation='sigmoid',kernel_initializer='uniform'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=100)
```

# predicting and Evaluation

```python
y_pred_test=ann.predict(x_test)
y_pred_train=ann.predict(x_train)
y_pred_test=(y_pred_test>=0.5)
y_pred_train=(y_pred_train>=0.5)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print('test_accuracy',accuracy_score(y_test,y_pred_test))
confusion_matrix(y_test,y_pred_test)
```

```python
print('test_accuracy',accuracy_score(y_train,y_pred_train))
```

```python
def classifier_built():
    classifier=Sequential()
    classifier=tf.keras.models.Sequential()
    classifier.add(Dense(input_dim=11,units=6,activation='relu',kernel_initializer='uniform'))
    # input_dim is the input that we have
    # units is the no of the hidden layer 
    # actiation activation function that we want to use
    # Dense:-are the lines that we connect from the input layers to the hidden layer
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(units=6,activation='relu',kernel_initializer='uniform'))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(units=1,activation='sigmoid',kernel_initializer='uniform'))
    classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
```

```python
from scikeras.wrappers import KerasClassifier
classifier = KerasClassifier(classifier_built,batch_size = 32,epochs = 100)
```

```python
from sklearn.model_selection import cross_val_score
acc=cross_val_score(classifier,x,y,cv=5)
```

# with the help of Hyper-parameter tuning we have to decide how many layers we hae to take ,which activation function will be best ,which normalisation will be better and which is the best optimizer,which is the best batch size

```python
estimator=KerasClassifier(classifier_built())
param_grid={'batch_size':[10,32],'epochs':[50,100],"optimizers":['adam','rmsprop']}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(param_grid,estimator,scoring='accuracy',cv=5)
grid_result=grid.fit(x_train,y_train)
```
